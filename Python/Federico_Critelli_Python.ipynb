## Funzione `riprova()`

Questa funzione chiede all'utente se vuole riprovare un'operazione.  

- L'utente deve rispondere `"si"` o `"no"`.  
- Se l'utente digita un valore non valido, viene mostrato un messaggio di errore e l'input viene richiesto di nuovo.  
- La funzione restituisce:
  - `True` se l'utente vuole riprovare (`"si"`)  
  - `False` se l'utente non vuole riprovare (`"no"`)  


```python
#funzione che chiede all'utente se vuole rirpovare un'operazione
def riprova():
    """
    Chiede all'utente se vuole riprovare un'operazione.

    Continua a chiedere finché l'utente non inserisce una risposta valida:
    'si' per riprovare, 'no' per non riprovare.

    Returns:
        bool: True se l'utente vuole riprovare,  altrimenti False.
    
    Esempio di utilizzo:
        if riprova():
            print("L'utente vuole riprovare.")
        else:
            print("L'utente non vuole riprovare.")
    """
    
    #chiede all'utente se vuole riprovare
    risposta = input("Vuole riprovare? 'si' o 'no': ").strip().lower()
    
    #continua a chiedere all'utente finchè la risposta non è valida
    while risposta not in ["si", "no"]:
        
        #messaggio di errore e nuova richiesta di input
        risposta = input("ERRORE: Digiti solamente 'si' o 'no': ").strip().lower()

    #restituisce la risposta
    return risposta == "si"
```

## Funzione `visualizzazione_viaggi()`

Questa funzione mostra in modo **sintetico** tutti i viaggi disponibili, indicando:  
continente: paese,città


### Come funziona:

- Raggruppa i viaggi in base al continente (Asia e America).  
- Scorre tutti i paesi di ciascun continente.  
- Scorre tutte le città disponibili in ogni paese.  
- Stampa solo le informazioni essenziali in formato leggibile.

### Nell'output sottostante si può osservare un esempio di funzionamento:

```python
def visualizzazione_viaggi(viaggi_asia, viaggi_america):
    """
    Mostra in modo sintetico i viaggi disponibili, indicando solo:
    continente : paese, città

    Args:
        viaggi_asia (dict): Dizionario dei viaggi disponibili in Asia.
        viaggi_america (dict): Dizionario dei viaggi disponibili in America.
    
    Returns:
        None
    
    """

    # Raggruppa i viaggi in un unico dizionario
    tutti_viaggi = {
        "Asia": viaggi_asia,
        "America": viaggi_america
    }
    
    #cicla su tutti i continenti
    for continente, paesi in tutti_viaggi.items():
       
        #cicla su tutti i paesi presenti nel continente
        for paese, info_paese in paesi.items():
           
            #cicla su tutte le città cisponibili nel paese
            for citta in info_paese["destinazioni"].keys():
              
                # Stampa solo continente, paese e città
                print(f"{continente.lower()} : {paese.lower()}, {citta.replace('_', ' ').capitalize()}")


#la funzione viene richiamata e mostra l'elenco di tutti i  viaggi contenuti nel dizionario
visualizzazione_viaggi(viaggi_asia, viaggi_america)
```

# Funzione `filtro`

La funzione `filtro` permette di **selezionare viaggi disponibili in Asia o America** in base al **budget dell'utente**.  

Dopo aver scelto il continente e inserito il budget massimo, l’utente visualizza le destinazioni compatibili e può generare un **preventivo personalizzato** indicando il numero di persone e la durata del viaggio.  

Gestisce automaticamente **errori di input** e assicura che le scelte siano valide prima di calcolare e restituire il preventivo finale.

### Nell'output sottostante si può osservare un esempio di funzionamento:


```python
#definisco la funzione filtro servirà a filtrare i possibili viaggi che rientrano nel budget dell'utente
def filtro(viaggi_asia, viaggi_america):
    """
    Filtra i viaggi disponibili in base al budget dell'utente e consente di
    generare un preventivo personalizzato.

    La funzione calcola dinamicamente il budget minimo e massimo dai prezzi
    presenti nei dizionari dei viaggi. L'utente può quindi scegliere il
    budget desiderato, visualizzare le destinazioni disponibili e richiedere
    un preventivo finale.

    Gestione errori:

        - Budget non numerico o fuori dal range disponibile
        - Città non presente nelle opzioni
        - Numero di persone o durata non valida

    Args:
        viaggi_asia (dict): Dizionario dei viaggi disponibili in Asia.
        viaggi_america (dict): Dizionario dei viaggi disponibili in America.

    returns:
        none or dict
        
    Esempio di utilizzo:
        filtro(viaggi_asia, viaggi_america)
    """

    
    #raggruppa i viaggi in un unico dizionario
    tutti_viaggi = {
        "asia":viaggi_asia,
        "america":viaggi_america
    }
    #variabile di controllo
    continente_valido = False
   
    #ciclo finchè l'utente non inserisce una scelta corretta
    while not continente_valido:
        #chiede all'utente in quale continente vuole viaggiare
        continente_scelto = input("In quale continente vuole viaggiare? 'Asia' o 'America': ").strip().lower()
        
        #controlla se l'input è uno dei continenti disponibili
        if continente_scelto in ["asia", "america"]:
            
            #se la scelta è valida, esce dal ciclo
            continente_valido = True
            
            # Filtra i viaggi solo sul continente scelto
            viaggi_selezionati = {continente_scelto: tutti_viaggi[continente_scelto]}
        
        #se l'input non è valido
        else:
            #mostra messaggio di errore
            print("\nERRORE: digitare solo 'Asia' o 'America'.")
            
            #chiede all'utente se vuole riprovare se no termina la funzione
            if not riprova():
                return
            

     #costruzione della lista completa di tutti i viaggi
    prezzi = []
    for continente, paesi in viaggi_selezionati.items():
        for paese, info_paese in paesi.items():
            for citta, info in info_paese["destinazioni"].items():
                prezzi.append(info["week1"])
                prezzi.append(info["week2"])
    
    #se non ci sono viaggi disponibili interrompe la funzione
    if not prezzi:
        print("\nNon ci sono viaggi disponibili.")
   
        return
   
    #determina il budget minimoe massimo disponibile
    budget_min = min(prezzi)
    budget_max = max(prezzi)
    
    #ciclo principale per gestire l'interazione con l'utente
    while True:
        
        #richiesta del budget con gestione degli errori
        try:
            budget = int(input( f"Considerando che i nostri pacchetti hanno un 'prezzo minimo di {budget_min}\n e un prezzo massimo di {budget_max}', Digiti il budget massimo che desidera spendere(€):"))
           
        except ValueError:
            print("\nERRORE: Deve inserire un numero intero")
            if not riprova():
                return
            continue
        
        #controllo che il budget sia nel range consentito
        if budget < budget_min or budget > budget_max:
            print(f"\nERRORE: Il budget deve essere compreso tra {budget_min} e {budget_max} €")
            if not riprova():
                return
            continue
        
        #lista che conterrà tutti i viaggi disponibili con il budget
        viaggi_disponibili = []

        print("\nEcco i viaggi che rientrano nel suo budget(prezzo p.p.):\n")
        
        #ciclo che scorre tutti i viaggi disponibili nel continente scelto
        for continente, paesi in viaggi_selezionati.items():
            for  paese, infopaese in paesi.items():
                for citta, info in infopaese["destinazioni"].items():
                     
                     #se il prezzo per una settimana rientra nel budget lo aggiunge alla lista
                    if info["week1"] <= budget:
                        viaggi_disponibili.append((continente, paese,citta, 1, info["week1"] ))
                        
                        print(f"{paese} - {citta} - € {info['week1']} - 1 settimana")
                    
                    #se il prezzo per due settimane rientra nel budget lo aggiunge alla lista
                    if info["week2"] <= budget:
                        viaggi_disponibili.append((continente, paese,citta, 2, info["week2"] ))
                        
                        print(f"{paese} - {citta} - € {info['week2']} - 2 settimana")
       
        #se non ci sono viaggi per il suo budget chiede se vuole riprovare
        if not viaggi_disponibili:
            print("\nNessun viaggio disponibile per questo budget.")
            if not riprova():
               
                return
            
            continue

        #chiede all'utente di scegliere una città tra quelle disponibili
        scelta_citta = input("\nScriva il nome della città che le interessa: ").strip().lower()
        
        
        #ricerca del viaggio scelto
        trovato = None
        for v in viaggi_disponibili:
            continente, paese, citta, durata, prezzo = v
            if citta == scelta_citta:
                trovato = v
               
                break
        
        #se non trova la citta chiede se riprovare
        if not trovato:
            print("\nERRORE: Viaggio non trovato.")
            if not riprova():
                
                return
           
            continue

        #chiede all'utente se vuole fare un preventivo per quel viaggio
        fare_preventivo = input("Vuole fare un preventivo per questo viaggio? 'si':o 'no' ").strip().lower()
        
        
        #se l'utente non vuole il preventivo termina la funzione
        if fare_preventivo == "no":
            return
        
        #gestione risposta non valida
        if fare_preventivo != "si":
            print("\nERRORE: risposta non valida.")
            if not riprova():
               
                return
            
            continue
        
        #scompone i dati del viaggio selezionato
        continente, paese, citta, durata, prezzo = trovato
        
        #richiede numero di persone e gestisce l'errore
        while True:
            try:
                persone = int(input("Quante persone?: "))
               
                if persone <= 0:
                    raise ValueError
                
                break
            
            except ValueError:
                print("\nERRORE: Inserisca un numero valido di persone.")
               
                if not riprova():
                    
                    return
        
        #chiede la dura del viaggio e gestisce l'errore
        while True:
           
            try:
                durata = int(input("Vuole fare un viaggio di 1 o 2 settimane?: "))
                
                if durata not in [1, 2]:
                    raise ValueError
                
                break
            
            except ValueError:
                print("\nERRORE: Inserisci 1 o 2.")
                if not riprova():
                    return
        
        #recupera informazioni aggiornate dal dizionario
        info = tutti_viaggi[continente][paese]["destinazioni"][citta]
        prezzo = info[f"week{durata}"]
       
        #calcolo totale
        totale = prezzo * persone

        #crea dizionario preventivo
        preventivo = {
            "continente": continente,
            "paese": paese,
            "citta": citta,
            "durata_settimane": durata,
            "prezzo_base": prezzo,
            "persone": persone,
            "totale": totale
        }

        #stampa preventivo finale
        print("\nPreventivo: \n")
        for k, v in preventivo.items():
             
             print(f"{k.capitalize()}: {v}")
       
        #esce dalla funzione dopo aver stampato il preventivo,impedisce che il ciclo while ricominci
        return preventivo
    
#richiama la funzione filtro
filtro(viaggi_asia, viaggi_america)    
```

## Funzione `lascia_valutazione()`

Questa funzione permette all'utente di lasciare una valutazione dell'esperienza su Orizon.  
L'utente può scegliere se lasciare una valutazione da 1 a 5:

- Se l'utente inserisce un valore valido, il punteggio viene aggiunto alla lista delle valutazioni.  
- Se l'utente inserisce un valore non valido (ad esempio lettere o numeri fuori dal range), viene mostrato un messaggio di errore e l'input viene richiesto di nuovo.  
- Se l'utente sceglie di non lasciare una valutazione, viene mostrato un messaggio di ringraziamento.

### Nell'output sottostante si può osservare un esempio di funzionamento:

```python
def lascia_valutazione(valutazioni):
    """
    Permette all'utente di lasciare una valutazione dell'esperienza su Orizon.

    La funzione richiede all'utente se desidera inserire una valutazione da 1 a 5.
    Se l'utente inserisce un valore valido, la valutazione viene aggiunta alla lista
    fornita come parametro.

    Args:
        valutazioni (list): Lista delle valutazioni esistenti che verrà aggiornata
            con la nuova valutazione inserita dall'utente.

    Returns:
        list: Lista aggiornata delle valutazioni.
    """

    # variabile di controllo per il ciclo while
    scelta_valida = False
    while not scelta_valida:
        #input di scelta
        scelta = input("\nVuole lasciare una valutazione dell'esperienza su Orizon? 'si' o 'no': ").strip().lower()
       
        if scelta in ["si", "no"]:

            #input valido, esce dal ciclo
            scelta_valida = True
        else:
            #input non valido, ripete il ciclo
            print("Errore: digitare solo 'si' o 'no'")

    # se l'utente vuole lasciare una valutazione
    if scelta == "si":
       
        # variabile di controllo punteggio
        sc_ok = False  
        
        #ciclo finchè l'utente non inserisce un punteggio valido
        while not sc_ok:
            punteggio = input("Dà una recensione da 1 a 5: ")

            try:
                #prova a convertire l'input in numero intero
                punteggio = int(punteggio)

                #verifica che il punteggio sia compreso tra 1 e 5
                if 1 <= punteggio <= 5:
                    
                    #aggiunge il punteggio alla lista
                    valutazioni.append(punteggio)
                    
                    #punteggio valido esce dal ciclo
                    sc_ok = True
                    
                    #messaggio di conferma
                    print("Recensione aggiunta correttamente")
                    #mostra il punteggio inserito
                    print(f"Hai inserito: {punteggio}")
                
                else:
                    #punteggio fuori range
                    print("Errore: digitare un numero tra 1 e 5")

            except ValueError:
                #gestione errore: l'input non era un numero
                print("Errore: digitare un numero tra 1 e 5")
   
    #se l'utente non vuole lasciare una recensione
    else:
        print("Grazie di aver scelto Orizon")

    #restituisce la lista aggiornata delle valutazioni
    return valutazioni

#chiamata della funzione passando la lista 'valutazioni' già definita
lascia_valutazione(valutazioni)
```

## Funzione `mostra_statistiche_complete()`

Questa funzione mostra un riepilogo completo delle statistiche di **Orizon Viaggi**.

### Cosa fa la funzione:

- Calcola il numero totale di valutazioni lasciate dagli utenti  
- Calcola la valutazione media (arrotondata a 2 decimali)  
- Mostra il numero di città disponibili per ogni paese  
- Mostra il totale dei viaggi disponibili per ciascun continente  

La funzione non restituisce valori ma stampa direttamente a schermo il riepilogo statistico.

### Nell'output sottostante si può osservare un esempio di funzionamento:


```python
#mostra tutte le statistiche disponibili
def mostra_statistiche_complete(valutazioni, viaggi_asia, viaggi_america):
    """
    Mostra statistiche generali di Orizon Viaggi, comprese valutazioni degli utenti
    e numero di viaggi disponibili per continente e per paese.

    La funzione stampa:
        - Numero totale di valutazioni lasciate dagli utenti
        - Valutazione media
        - Numero di città disponibili per ogni paese
        - Totale viaggi disponibili per continente

    Args:
        valutazioni (list): Lista delle valutazioni già registrate.
        viaggi_asia (dict): Dizionario dei viaggi disponibili in Asia.
        viaggi_america (dict): Dizionario dei viaggi disponibili in America.

    returns:
        none
    
    Esempio di utilizzo:
        mostra_statistiche_complete(valutazioni, viaggi_asia, viaggi_america)
    """
    
    #titolo
    print("\nSTATISTICHE ORIZON\n")

    # verifica se esistono valutazioni salvate
    if valutazioni:
       
        #calcola il numero dei commenti
        totale_commenti = len(valutazioni)
        
        #calcola la media e arrotonda a 2 decimali
        media = round(sum(valutazioni) / totale_commenti, 2)
    
    else:
        #se non ci sono valutazioni imposta 0 di default
        totale_commenti = 0
        media = 0
    
    #stampa statistiche sulle valutazioni
    print(f"Numero valutazioni lasciate: {totale_commenti}")
    print(f"Valutazione media: {media} / 5\n")
    
    #ragruppa tutti i viaggi per continente in un unico dizionario
    tutti_viaggi = {
        "Asia": viaggi_asia,
        "America": viaggi_america
    }

    #ciclo sui continenti disponibili
    for continente, paesi in tutti_viaggi.items():
       
        #contatore numero tot. viaggi del continente
        num_viaggi = 0
        #stampa nome del continente
        print(f"\n{continente}")
        
        #ciclo sui paesi del continente
        for paese, info in paesi.items():
            
            #calcola il numero di città disponibili per il paese
            num_citta = len(info["destinazioni"])
            
            #aggiorna il totale viaggi del continente
            num_viaggi += num_citta
            
            #stampa il numero di città per paese
            print(f"  - {paese}: {num_citta} città")
        
        #stampa il totale di viaggi disponibili per continente
        print(f"Totale viaggi in {continente}: {num_viaggi}")

#richiama la funzione
mostra_statistiche_complete(valutazioni, viaggi_asia, viaggi_america)
```

# Funzione: cerca_viaggio

Questa funzione permette all'utente di cercare una **città specifica** tra i viaggi disponibili in **Asia** e **America**.  

**Caratteristiche principali:**

- L'utente inserisce il nome della città desiderata.
- La funzione verifica se la città esiste nei dizionari dei viaggi.
- Se la città è trovata, stampa tutte le informazioni riguardati la città selezionata.
- Gestisce gli errori se l'utente inserisce **numeri** o valori non validi, usando `try/except` e la funzione `riprova()`.
- La funzione mantiene un ciclo finché l'utente non inserisce una città valida o decide di uscire.

### Nell'output sottostante si può osservare un esempio di funzionamento:

```python
def cerca_viaggio(viaggi_asia, viaggi_america):
    """
    Permette all'utente di cercare una città tra i viaggi disponibili in Asia e America.
    Se la città esiste, stampa continente, paese, città, prezzi per 1 e 2 settimane e tipologia.
    Gestisce errori se l'utente inserisce numeri o valori non validi.

    Args:
        viaggi_asia (dict): Dizionario dei viaggi disponibili in Asia.
        viaggi_america (dict): Dizionario dei viaggi disponibili in America.

    Returns:
        None
    """
    
    # crea un set con tutte le tipologie/città disponibili nei due dizionari
    # crea un set con tutte le città disponibili nei due dizionari
    citta_disponibili = set(
        [nome_citta
        for paese in viaggi_asia.values()
        for nome_citta in paese["destinazioni"].keys()] +
        [nome_citta
        for paese in viaggi_america.values()
        for nome_citta in paese["destinazioni"].keys()]
)

    
    # unisce tutti i viaggi in un dizionario
    tutti_viaggi = {
        "asia": viaggi_asia,
        "america": viaggi_america
    }
    
    while True:
        citta = input(f"Digiti il nome della città in cui vuole viaggiare: {citta_disponibili}").strip().lower()
        
        # gestione errore se l'utente inserisce un numero
        try:
            if citta.isnumeric():
                raise ValueError("\nNon può inserire numeri come nome della città.")
       
        except ValueError as e:
            print(f"ERRORE: {e}")
            if not riprova():
                return
            continue
        
        # ricerca della città nei due dizionari
        trovato = False
        for continente, paesi in tutti_viaggi.items():
            for paese, info_paese in paesi.items():
                for nome_citta, info_citta in info_paese["destinazioni"].items():
                   
                    #confronta la città inserita con quelle presenti
                    if nome_citta == citta:
                        
                        #se la città viene trovata aggiorna il flag
                        trovato = True
                        
                        # stampa tutte le info relative alla città trovata
                        print(f"\nCittà trovata.\n")
                        print(f"Continente: {continente.capitalize()}")
                        print(f"Paese: {paese.capitalize()}")
                        print(f"Città: {nome_citta.capitalize()}")
                        print(f"Prezzo 1 settimana: €{info_citta['week1']}")
                        print(f"Prezzo 2 settimane: €{info_citta['week2']}")
                        print(f"Tipologia: {info_citta['tipologia'].capitalize()}\n")
                        break
               #se la città è stata trovata esce dal ciclo dei paesi
                if trovato:
                    break
           
           #se la città è stata trovata esce dal ciclo dei continenti
            if trovato:
                break
        #se la città non è stata trovata
        if not trovato:
            print("Città non trovata.\n")
           
            #chiede se l'utente vuole riprovare
            if not riprova():
                return
        
        # esce dal ciclo se la città è stata trovata e visualizzata
        if trovato:
            break

#richiama la funzione
cerca_viaggio(viaggi_asia, viaggi_america)
```

# Funzione `aggiungi_viaggio ()`

Questa funzione permette all’amministratore di aggiungere un nuovo viaggio nei dizionari dei viaggi disponibili, suddivisi per continente (Asia o America).  

## Cosa fa
- Richiede all’utente di inserire continente, paese, città, prezzi per 1 e 2 settimane e tipologia del viaggio.
- Gestisce errori di input per continente e prezzi, permettendo di riprovare.
- Controlla se la città esiste già e chiede conferma per sovrascrivere il viaggio esistente.
- Aggiunge i dati del nuovo viaggio al dizionario corrispondente.
- Stampa un riepilogo del viaggio appena aggiunto e conferma l’avvenuto inserimento.

## Valore restituito
- La funzione non restituisce valori (`None`). Termina dopo l’aggiunta del viaggio o in caso di errori.

### Nell'output sottostante si può osservare un esempio di funzionamento:

```python
def aggiungi_viaggio(viaggi_asia, viaggi_america):
    """
    Permette l'aggiunta di un nuovo viaggio nei dizionari
    dei viaggi disponibili (Asia o America).

    Args:
        viaggi_asia (dict): Dizionario dei viaggi disponibili in Asia.
        viaggi_america (dict): Dizionario dei viaggi disponibili in America.

    Returns:
        None

    Esempio:
        aggiungi_viaggio(viaggi_asia, viaggi_america)    
    """
     #ciclo principale per aggiungere un viaggio
    while True:
      
        #ciclo per chiedere il continente finchè l'imput non è valido
        while True:
            continente = input("Continente (Asia/America): ").strip().lower()
        
            if continente == "asia":
                viaggi = viaggi_asia
                #continente corretto esce dal ciclo
                break 
        
            elif continente == "america":
                viaggi = viaggi_america
                #continente correto esce dal ciclo
                break 
            
            else:
                #messaggio per continente non valido
                print("Continente non valido. Orizon organizza viaggi solamente in Asia o America")
                #se l'utente non vuole riprovare esce dalla funzione
                if not riprova():
                    return
        
        #Crea un set con tutte le tipologie già presenti
        #Serve come suggerimento per l'inserimento della nuova tipologia
        tipologie = set([destinazione["tipologia"] for paese in viaggi.values() for destinazione in paese["destinazioni"].values()])
       
        #richiesta input paese e città    
        paese = input("Paese: ").strip().lower()
        citta = input("Città: ").strip().lower()
        
        #ciclo per inserire i prezzi finchè l'input non è valido
        #inserimento prezzo prima settimana
        while True:
            try:
                week1 = int(input("Prezzo 1 settimana: "))
                break
            except ValueError:
                print("\nPrezzo non valido. Deve inserire un numero intero")
                if not riprova():
                    return
        
        #ciclo per inserire i prezzi finchè l'input non è valido
        #inserimento prezzo seconda settimana
        while True:
            try:
                week2 = int(input("Prezzo 2 settimane: "))
                break
            except ValueError:
                print("\nPrezzo non valido. Deve inserire un numero intero")
                if not riprova():
                    return

       
        #richiesta input tipologia del viaggio
        tipologia = input(f"Tipologia (es:{tipologie}): ").strip().lower()
        
        #se il paese ancora non esiste nel dizionario lo crea
        if paese not in viaggi:
            viaggi[paese] = {"paese": paese, "destinazioni": {}}
        
        #controllo se la città esiste già
        if citta in viaggi[paese]["destinazioni"]:
            print(f"\nAttenzione: il viaggio per {citta.capitalize()} esiste già.")
            sovrascrivi = input("Vuoi sovrascriverlo? ('si' o 'no'): ").strip().lower()
            if sovrascrivi != "si":
                print("\nIl viaggio non è stato sovrascritto.")
                #ritorna al ciclo principale
                continue
        
        
        
        #aggiunge la nuova città e i relativi dati nel dizionario
        viaggi[paese]["destinazioni"][citta] = {
            "week1": week1,
            "week2": week2,
            "tipologia": tipologia
        }
        
        #stampa un riepilogo del viaggio appena aggiunto con nome città, paese e continente
        print(f"\n{citta.capitalize()} {paese.capitalize()}, {continente.capitalize()}")
       
        #mostra i prezzi per 1 e 2 settimane e la tipologia del viaggio
        print(f"\n1 settimana: €{week1}, 2 settimane: €{week2}, Tipologia: {tipologia}\n")
       
        #messaggio di conferma che il viaggio è stato aggiunto correttamente
        print("\nVIAGGIO AGGIUNTO CORRETTAMENTE.")
        
        #termina la funzione dopo aver completato l'inserimento del viaggio
        return

#richiama la funzione
aggiungi_viaggio(viaggi_asia, viaggi_america)    
```

# Funzione `orientamento ()`

La funzione `orientamento` guida l'utente nella scelta del viaggio ideale tra le destinazioni disponibili in Asia o America.  

Permette di selezionare:
- Il continente (Asia o America)
- Il paese
- La città
- La durata del viaggio (1 o 2 settimane)
- Il numero di persone  

Calcola il preventivo personalizzato e offre all’utente la possibilità di vedere altri viaggi che rientrano nel suo budget tramite la funzione `filtro()`.  

**Valore restituito:**  
Può restituire il preventivo generato da `filtro()`, una stringa di conferma oppure `None`, a seconda della scelta dell’utente.

### Nell'output sottostante si può osservare un esempio di funzionamento:

```python
# funzione di orientamento nella scelta del viaggio
def orientamento(viaggi_asia, viaggi_america):
    """
    Guida l'utente nella scelta del viaggio ideale mostrando paesi e destinazioni disponibili.
    
    L'utente può selezionare:
        - Il continente (Asia o America)
        - Il paese
        - La città
        - La durata del viaggio (1 o 2 settimane)
        - Il numero di persone

    La funzione calcola il preventivo e chiede se l'utente vuole vedere altri viaggi che rientrano nel suo budget.

    Args:
        viaggi_asia (dict): Dizionario contenente i viaggi disponibili in Asia.
        viaggi_america (dict): Dizionario contenente i viaggi disponibili in America.

    Returns:
        La funzione può restituire il risultato della funzione filtro(),
        una stringa di conferma oppure none, a seconda della scelta dell'utente.
    """

    
    #ciclo infinito continua finchè l'utente non decide di uscire
    while True:
        
        #richiesta di inserire il continente che si preferisce
        continente = input("La nostra agenzia organizza viaggi in Asia e America. Digiti 'Asia' o 'America' per ricevere il piano più adatto a lei: ").strip().lower()
        
        
        #se il continente scelto è Asia assegna il dizionario viaggi_asia al dizionario viaggi
        if continente == "asia":
            viaggi = viaggi_asia
       
        #se il continente scelto è America assegna il dizionario viaggi_america al dizionario viaggi
        elif continente == "america":
            viaggi = viaggi_america
        
        #se il continente non è valido chiede al cliente se vuole ricominciare se si ricomincia altrimenti esce dal ciclo
        else:
            print("Continente non valido.")
            if riprova():
                continue
            else:
                break
        
        #stampa le destinazioni disponibili per il continente scelto
        print("Questi sono i paesi disponibili per il continente da lei scelto:")
        
        for k in viaggi:
            print("-", k)
        
        #chiede all'utente di scegliere il paese
        scelta_paese = input("In quale paese vuole viaggiare? ").strip().lower()
        
        
        #verifica la validità del paese
        if scelta_paese in viaggi:
            
            #stampa le destinazioni disponibili per il paese scelto
            print("\nOra può scegliere tra queste destinazioni:")
            for destinazione, info in viaggi[scelta_paese]["destinazioni"].items():
                print(f"{destinazione} - {info['tipologia']}")
           
            #chiede la citta scelta all'utente
            citta = input("Inserisca la città che preferisce così da poter strutturare un preventivo su misura: ").strip().lower()
            
           
            #controlla se la città esiste nel dizionario
            if citta in viaggi[scelta_paese]["destinazioni"]:
                
                #richiesta durata viaggio
                durata = int(input("Per quante settimane vuole partire? '1' o '2'): "))
                
                
                #verifica che la durata sia valida
                if durata in [1, 2]:
                   
                    #informazioni sulla destinazione scelta
                    info = viaggi[scelta_paese]["destinazioni"][citta]
                    prezzo = info[f"week{durata}"]
                    
                    #chiede il numero di persone
                    persone = int(input("Per quante persone desidera il preventivo?: "))
                   
                   
                    #calcola il prezzo totale
                    totale = prezzo * persone
                   
                    #stampa preventivo
                    print(f"\nPreventivo per {citta.capitalize()}:")
                    print(f"Tipologia: {info['tipologia']}")
                    print(f"Durata: {durata} settimana/e")
                    print(f"Prezzo base: {prezzo} €")
                    print(f"Numero persone: {persone}")
                    print(f"Prezzo totale: {totale} €")
                   
                    #ciclo per chiede al cliente se vuole vedere i viaggi che rientrano nel suo budget
                    while True:
                        scelta_cliente = input(("se vuole scoprire tutti i viaggi che rientrano nel suo budget digiti 'si' altrimenti 'no'")).strip().lower()
                        
                       
                       #se l'utente risponde si chiama la funzione filtro() per mostrare i viaggi disponibili
                        if scelta_cliente == 'si':
                            return filtro(viaggi_asia, viaggi_america)
                        
                        #se l'utente non vuole continuare
                        elif scelta_cliente == "no":
                            return "Grazie per consultato Orizon viaggi"

                        #gestione risposta non valida
                        else:
                            print("ERRORE: Digiti solamente 'si' o 'no'")

                #durata non valida
                else:
                    print("Durata non valida.")
                    riprovare = input("ERRORE: Vuole riprovare? 'si' o 'no': ").strip().lower()
                    
                    if riprova():
                        continue
                    else:
                        break
           
            #citta non valida
            else:
                print("Città non valida.")
                if riprova():
                    continue
                else:
                    break
       
        #paese non valido
        else:
            print("Paese non valido.")
            if riprova():
                continue
            else:
                break

#richiama la funzione
orientamento(viaggi_asia, viaggi_america)
```

Questa funzione gestisce il **menù principale** dell’agenzia di viaggi Orizon, permettendo all’utente di interagire con tutte le principali funzionalità disponibili.  

### Funzionalità principali:
- Guidare l’utente nella scelta del viaggio ideale (**orientamento**)
- Cercare viaggi in base al **budget** (**filtro**)
- Visualizzare le **statistiche dell’agenzia** (**mostra_statistiche_complete**)

- Aggiungere nuovi viaggi (**aggiungi_viaggio**) 
- Cercare un viaggio digitando la **città** (**cerca_viaggio**)
- Lasciare una **valutazione** (**lascia_valutazione**)

### Caratteristiche della funzione:
- Mantiene un ciclo attivo finché l’utente non decide di uscire
- Gestisce gli input non validi con messaggi di errore chiari
- Richiama le funzioni corrispondenti in base alla scelta dell’utente
- Visualizza logo e messaggio di benvenuto all’avvio

### Input richiesti:
- `viaggi_asia` (dict): dizionario dei viaggi disponibili in Asia
- `viaggi_america` (dict): dizionario dei viaggi disponibili in America
- `valutazioni` (list): lista delle valutazioni lasciate dagli utenti
- `logo` (str): logo dell’agenzia
- `benvenuto` (str): messaggio di benvenuto per l’utente

### Output:
- Nessun valore restituito (`None`), ma l’utente interagisce direttamente tramite input e output a schermo.

Questa funzione costituisce il **punto di accesso principale** per tutte le funzionalità offerte dall’agenzia.

### Nell'output sottostante si può osservare un esempio di funzionamento:

```python
#gestisce il menù principale
def menu_principale(viaggi_asia, viaggi_america, valutazioni, logo, benvenuto):
    """
    Mostra il menù principale di Orizon Viaggi.
    
    Args:
        viaggi_asia (dict): Dizionario dei viaggi disponibili in Asia.
        viaggi_america (dict): Dizionario dei viaggi disponibili in America.
        valutazioni (list): Lista delle valutazioni degli utenti.
        logo (str): Logo da mostrare all'avvio.
        benvenuto (str): Messaggio di benvenuto.
        
    Returns:
        None
    """

    #ciclo infinito per mantenere attivo il menù finchè l'utente non sceglie di uscire
    while True:
        #stampa logo e messaggio di benvenuto
        print(logo)
        print(benvenuto)
        
        #stampa opzioni disponibili
        print("\nCosa desidera fare?")
        print("\n---FUNZIONI BASE---\n")
        print("1 - Vuoi aggiungere un nuovo viaggio?")
        print("2 - Visualizzare tutti i viaggi disponibili")
        print("3 - Cercare un viaggio digitando la città che più le interessa")
        print("\n---FUNZIONALITÀ AGGIUNTIVE---\n")
        print("4 - Visualizzare le statistiche Orizon")
        print("5 - Cercare un viaggio in base al budget")
        print("\n---EXTRA---\n")
        print("6 - Essere guidato nella scelta del viaggio ideale")
        print("7 - Lasciare una valutazione")
        print("NULLA - Premi INVIO per uscire\n")
       
        #inserimento scelta del cliente ed elimina eventuali spazi
        scelta = input("Selezioni un'opzione (1-7): ").strip()
        
        
       
        
        #se l'utente sceglie 1 avvia la funzione aggiungi_viaggio
        if scelta == "1":
            aggiungi_viaggio(viaggi_asia, viaggi_america)
        
        #se sceglie 2 avvia la funzione visualizzazione_viaggi
        elif scelta == "2":
            visualizzazione_viaggi(viaggi_asia, viaggi_america)
        
        #se sceglie 3 avvia la funzione mostra cerca_viaggio
        elif scelta == "3":
            cerca_viaggio()
        
        #se sceglie 4 avvia la funzione mostra_statistiche_complete
        elif scelta =="4":
            mostra_statistiche_complete(valutazioni, viaggi_asia, viaggi_america)

        #se l'utente digita 5 avvia la funzione filtro
        elif scelta == "5":
                filtro(viaggi_asia, viaggi_america)
       
        #se l'utente digita 6 avvia la funzione orientamento
        elif scelta == "6":
            orientamento(viaggi_asia, viaggi_america)
            
        
        #se l'utente digita 7 avvia la funzione lascia_valutazione
        elif scelta == "7":
            lascia_valutazione(valutazioni)
        
        #se l'utente preme invio esce dal menù
        elif scelta == "":
            break
        
        #gestione inserimenti non validi
        else:
            print("\nERRORE: scelta non valida. Inserisca un numero intero da '1' a '5'")

#avvia il menù principale del programma, mostra le opzioni disponibili all'utente 
#finchè l'utente non decide di uscire
menu_principale(viaggi_asia, viaggi_america, valutazioni, logo, benvenuto)
```
